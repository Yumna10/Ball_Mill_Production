# Cleaning & Transformation Def

```python
def clean_data(file_path):
  df = pd.read_excel(file_path)
  date_col = df['Unnamed: 1'].iloc[1]
  df['Date'] = date_col
  new_header = df.iloc[3]
  df = df[4:]
  df.columns = new_header
  df = df.reset_index(drop = True)
  col_names = ['Ball Mill', 'Limestone', 'Shale', 'Sand', 'Iron Ore', 'Total Production', 'Date']
  df.columns = col_names
  df[['Limestone', 'Shale', 'Sand', 'Iron Ore', 'Total Production']] = df[['Limestone', 'Shale', 'Sand', 'Iron Ore', 'Total Production']].astype(int)
  return df
```

```python
folder_path = '/content/drive/MyDrive/Data/Cleaning Data - Practice'
dfs = []

for file_name in os.listdir(folder_path):
  file_path = os.path.join(folder_path, file_name)
  cleaned_df = clean_data(file_path)
  if cleaned_df is not None:
    dfs.append(cleaned_df)

final_df = pd.concat(dfs, ignore_index=True)
final_df
```

# Total Production in each Ball Mill across time

```python
pd.pivot_table(
    data = final_df,
    index = 'Date',
    columns = 'Ball Mill',
    values = 'Total Production'
)
```

```python
pd.pivot_table(
    data = final_df,
    index = ['Limestone','Shale', 'Sand','Iron Ore'],
    columns = 'Date',
    values = 'Total Production',
    aggfunc = 'sum'
)
```

```python
pd.pivot_table(
    data = final_df,
    columns = 'Ball Mill',
    values=['Limestone', 'Shale', 'Sand', 'Iron Ore'],
    index='Date',
    aggfunc='sum'
    )
```
